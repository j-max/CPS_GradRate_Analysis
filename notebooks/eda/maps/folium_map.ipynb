```python
import folium
import json
```

To make a folium map, I'll start with this [quickstart guide](https://python-visualization.github.io/folium/quickstart.html#Getting-Started)

```python
# map centered on Chicago
m = folium.Map(location=[41.881832,-87.623177])

```

```python
with open('d3_map/map_html/school_locs.geojson', 'rb') as read_file:
    schools = json.load(read_file)
```

```python
schools
```

```python
schools['features'][0]['geometry']['coordinates'][0]
```

```python
for school in schools['features']:
    if school['properties']['grade_cat'] == 'HS':
        folium.Marker((school['geometry']['coordinates'][1], school['geometry']['coordinates'][0]),
                      popup=school['properties']['name'],
                      icon=folium.Icon(color='red',icon_color='#FFFF00')).add_to(m)
    else:
        folium.Marker((school['geometry']['coordinates'][1], school['geometry']['coordinates'][0]),
                      popup=school['properties']['name'],
                      icon=folium.Icon(color='blue',icon_color='#FFFF00')).add_to(m)
        
```

```python
m
```

```python
folium.GeoJson('d3_map/map_html/school_locs.geojson', style_function=lambda feature: {
        'fillColor': '#ffff00',
        'color': 'black',
        'weight': 2,
        'dashArray': '5, 5'
    }).add_to(m)
```
