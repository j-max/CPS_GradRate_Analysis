```python
import math
df_nonan_zip = df.dropna(subset = ['Zip'])
for i, row in df.iterrows():
    if math.isnan(row['Zip']):
        correct_zip = df_nonan_zip.loc[df_nonan_zip.School_ID == row['School_ID']]['Zip']
    df.set_value(i, 'Zip', correct_zip)


        
        #print(row[['School_ID', 'Zip']])
```

```python
#fill in zipcodes
df_zips = df[['School_ID', 'Zip']]
#drop na
df_zips = df_zips.dropna(subset = ['Zip'])
df_zips = df_zips.drop_duplicates(subset = ['Zip'])
#create zip dictionary with school it
zip_dict = df_zips.to_dict()
#zip_dict[11].dtype
```

```python
for key,value in df_zips.items():
    df.loc[df.School_ID == key, 'Zip'] = value
```
